```python
import numpy as np

def digits2int(digits):
    return int(''.join([str(x) for x in digits]))
```

```python
# Global variables.
num_symbols = 4    # ATCG
num_digits = 5    # some integers comes from a string of the form 'X_1 X_2...X_n' where each X_i is one of the symbols (above) and has num_digits digits
num_digits_2 = 2
num_digits_3 = 6
num_digits_4 = 4
a_min, a_max = 4, 8
b_min, b_max = 4, 8
c_min, c_max = 4, 8
max_connections_input_neuron = 8
max_connections_output_neuron = 5
mapping = {0: 0.2, 1: 0.4, 2: 0.6, 3: 0.8}

# Dictionary containing all genes in symbolic form.
gens = {}

# Dictionary containing all genes in numeric form.
gens_num = {}

######################################################################################################
# Dimensions.
gens['dims'] = np.random.randint(num_symbols, size=[3, num_digits])

digits = gens['dims'][0, :]
a = max(a_min, digits2int(digits) % (a_max + 1))

digits = gens['dims'][1, :]
b = max(b_min, digits2int(digits) % (b_max + 1))

digits = gens['dims'][0, :]
c = max(c_min, digits2int(digits) % (c_max + 1))

gens_num['dims'] = (a, b, c)
num_neurons = a * b * c
print(f'Dimensions = {a} x {b} x {c}    Number of neurons = {num_neurons}')

######################################################################################################
# Connections with inputs.
gens['inputs'] = np.random.randint(num_symbols, size=[max_connections_input_neuron, 3 * num_digits])
gens_num['inputs'] = []

for i in range(max_connections_input_neuron):
    neuron_x = digits2int(gens['inputs'][i, : num_digits]) % (a + 1)
    neuron_y = digits2int(gens['inputs'][i, num_digits : 2*num_digits]) % (b + 1)
    neuron_z = digits2int(gens['inputs'][i, 2*num_digits : 3*num_digits]) % (c + 1)    
    gens_num['inputs'].append((neuron_x, neuron_y, neuron_z))

gens_num['inputs'] = list(set(gens_num['inputs']))
num_connections_input_neuron = len(gens_num['inputs'])
print(f"Neurons connected to inputs: {gens_num['inputs']}    Number of connections: {num_connections_input_neuron}")

######################################################################################################
# Connections with outputs.
gens['outputs'] = np.random.randint(num_symbols, size=[max_connections_output_neuron, 3 * num_digits])
gens_num['outputs'] = []

for i in range(max_connections_output_neuron):
    neuron_x = digits2int(gens['outputs'][i, : num_digits]) % (a + 1)
    neuron_y = digits2int(gens['outputs'][i, num_digits : 2*num_digits]) % (b + 1)
    neuron_z = digits2int(gens['outputs'][i, 2*num_digits : 3*num_digits]) % (c + 1)    
    gens_num['outputs'].append((neuron_x, neuron_y, neuron_z))

gens_num['outputs'] = list(set(gens_num['outputs']))
num_connections_output_neuron = len(gens_num['outputs'])

for x in gens_num['outputs']:
    if x in gens_num['inputs']:
        print(f'É necessário repetir a operação pois um neurônio conectado a output já estava sendo usado por um de input: {x}')

print(f"Neurons connected to outputs: {gens_num['outputs']}    Number of connections: {num_connections_output_neuron}")

######################################################################################################
# Region sizes.
gens['regions'] = np.random.randint(num_symbols, size=[num_neurons, 3 * num_digits_2])
gens_num['regions'] = {}
w = 0

for i in range(a):
    for j in range(b):
        for k in range(c):
            region_size_x = digits2int(gens['regions'][w, : num_digits_2]) % (a + 1)
            region_size_y = digits2int(gens['regions'][w, num_digits_2 : 2*num_digits_2]) % (b + 1)
            region_size_z = digits2int(gens['regions'][w, 2*num_digits_2 : 3*num_digits_2]) % (c + 1)    
            gens_num['regions'][(i, j, k)] = (region_size_x, region_size_y, region_size_z)
            w += 1

print(f"Region sizes of neuron (1, 1, 1) = {gens_num['regions'][(1, 1, 1)]}, ...")

######################################################################################################
# Probabilities of making connections.
gens['probs'] = np.random.randint(num_symbols, size=[num_neurons])
gens_num['probs'] = {}
w = 0

for i in range(a):
    for j in range(b):
        for k in range(c):
            gens_num['probs'][(i, j, k)] = mapping[gens['probs'][w]]
            w += 1

print(f"Connection probability of neuron (1, 1, 1) = {gens_num['probs'][(1, 1, 1)]}, ...")

######################################################################################################
# Threshold per neuron.
gens['thr'] = np.random.randint(num_symbols, size=[num_neurons])
gens_num['thr'] = {}
w = 0

for i in range(a):
    for j in range(b):
        for k in range(c):
            gens_num['thr'][(i, j, k)] = mapping[gens['thr'][w]]
            w += 1

print(f"Threshold of neuron (1, 1, 1) = {gens_num['thr'][(1, 1, 1)]}, ...")

######################################################################################################
# Decay of neuron connections.
gens['decay'] = np.random.randint(num_symbols, size=[num_digits_3])
gens_num['decay'] = float('0.' + ''.join([str(x) for x in gens['decay']]))
print(f"Decay e = {gens_num['decay']}")

# Minimum decay threshold to undo connection.
gens['decay_min'] = np.random.randint(num_symbols, size=[num_digits_4])
gens_num['decay_min'] = gens_num['decay'] / (digits2int(gens['decay_min']) + 1)
print(f"Decay minimum e_min = {gens_num['decay_min']}")

# number of iterations that the neuron is inactive after sending a signal.
gens['inactive_iter'] = np.random.randint(num_symbols, size=[num_digits_2])
gens_num['inactive_iter'] = digits2int(gens['inactive_iter']) + 1
print(f"Number of inactive iterations after sending a signal = {gens_num['inactive_iter']}")

# Number of iterations that the neuron is weakened after sending a signal.
gens['weakened_iter'] = np.random.randint(num_symbols, size=[num_digits_2])
gens_num['weakened_iter'] = gens_num['inactive_iter'] + digits2int(gens['weakened_iter']) + 1
print(f"Number of weakened iterations after sending a signal = {gens_num['weakened_iter']}")

# Weakening factor during iterations weakened.
gens['weakened_factor'] = np.random.randint(num_symbols, size=[num_digits_2])
gens_num['weakened_factor'] = float('0.' + ''.join([str(x) for x in gens['weakened_factor']]))
print(f"Weakening factor during iterations weakened = {gens_num['weakened_factor']}")
```
